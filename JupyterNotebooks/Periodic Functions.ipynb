```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plotF(func):
    X = np.linspace(-20, 20, 10000, endpoint=True)
    vfunc = np.vectorize(func)
    R = vfunc(X)

    plt.plot(X, R)
    plt.show()
```

# Periodic functions

Lets look at how we can make a periodic sawtooth function.

The **modulo** function can be used to map the reals or integers down to a quotent space. 

Here is an example, the intergers modulo 4 are $\{0,1,2,3\}$ expressed as $\mathbb{Z}/4\mathbb{Z}$ which is often notated as $\mathbb{Z}_4$

When modulo is done over the reals it produces a periodic function like those bellow. 

```python
def modulo(a):
    ''' R->R/10Z'''
    return a%10

plotF(modulo)
```

This makes a sawtooth function of period 10 and range [0,10]. If we want to make it into a sawtooth function of range [-1,1] we need to modify it

```python
def sawtooth(a):
    ''' Sawtooth function with period 10 and range -1 to 1 R->R'''
    x = a%10
    # It maps the Reals down to [0,10) 
    
    # need a line of y=mx+b that goes from -1 to 1
    # y = -1 at x=0 so b=  -1
    # y=1 at x=10  so 1 = m(10)-1    2=10m  m=1/5
    return 1/5.0*x-1.0

plotF(sawtooth)
```
